## Add the number of events to df_simple from sibil

```python
import pandas as pd
import pickle 
```

```python
df_simple = pickle.load(open('df_simple.p','rb'))
df_sibil = pd.read_csv('../dataset_demographie/sibil-systeme-d-information-billetterie.csv',sep=';')
```

```python
df_simple
```

```python
## Récupérer les représentations en 2024
start_date = pd.to_datetime("2024-01-01")
end_date = pd.to_datetime("2024-12-31")
df_sibil['declaration_date_representation'] = pd.to_datetime(df_sibil['declaration_date_representation'])
df_sibil_2024 = df_sibil[(df_sibil['declaration_date_representation']>=start_date)& (df_sibil['declaration_date_representation']<=end_date)]
df_sibil_2024
```

```python
df_sibil_relevant_information = df_sibil_2024[[ 'declarant_nom_usuel','declaration_date_representation'] ].groupby(['declarant_nom_usuel']).size().reset_index(name='nb_events_sibil')
```

```python
## faire un fuzzy join
import pandas as pd
from thefuzz import fuzz
from thefuzz import process

# Example DataFrames
df1 = pd.DataFrame({'a': ['Apple Inc.', 'Google LLC', 'Microsoft Corporation']})
df2 = pd.DataFrame({'b': ['Apple', 'Google', 'Microsoft', 'Amazon']})

def find_best_match(value, possible_matches, threshold=95):
    result = process.extractOne(value, possible_matches, scorer=fuzz.token_sort_ratio)
    if result:
        match, score = result[0], result[1]
        return match if score >= threshold else None
    return None


# Apply fuzzy matching to df1['a'] against df2['b']
df1['b_match'] = df1['a'].apply(lambda x: find_best_match(x, df2['b']))

# Perform a classic left join on the fuzzy-matched column
df_merged = df1.merge(df2, left_on='b_match', right_on='b', how='left')

print(df_merged)
```

```python
# Apply to df_simple and sibil
df_simple['name_match'] = df_simple['name'].apply(lambda x: find_best_match(x, df_sibil_relevant_information['declarant_nom_usuel']))

# Perform a classic left join on the fuzzy-matched column
df_merged = df_simple.merge(df_sibil_relevant_information, left_on='name_match', right_on='declarant_nom_usuel', how='left')
```
